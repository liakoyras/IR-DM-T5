```python
import os
import time

def time_extraction(fun,args):
    start = time.time()
    fun(args)
    return time.time()-start

def extract_la(doc_with_list):
    lines = []
    doc = []

    f = open(doc_with_list, "r")
    lines = lines + f.read().splitlines()

    for i in lines:
        f = open('latimes/'+i[:-5].lower(), 'r')
        text = f.read().splitlines()
        for j in range(len(text)):
            if i in text[j]:
                p=j-1
                while '</DOC>' not in text[p]:
                    doc.append(text[p])
                    p+=1
                doc.append('</DOC>')
                break

    with open('la_titles_only.result', 'w') as f:
        for item in doc:
            f.write("%s\n" % item)

            
def extract_ft(doc_with_list):
    lines = []
    doc = []

    f_text = []
    f_id = []

    for i in os.listdir('ft'):
        text = []
        for j in os.listdir('ft/'+i):    
            f = open('ft/'+i+'/'+j)
            text = text + f.read().splitlines()
        f_text.append(text)
        f_id.append(i)  

    f = open(doc_with_list, "r")
    lines = lines + f.read().splitlines()

    read_files=[]

    for i in lines:
        for j in range(len(f_text[f_id.index(i[:5].lower())])):
            if i in f_text[f_id.index(i[:5].lower())][j]:
                p=j-1
                while '</DOC>' not in f_text[f_id.index(i[:5].lower())][p]:
                    doc.append(f_text[f_id.index(i[:5].lower())][p])
                    p+=1
                doc.append('</DOC>')
                break

    with open('ft_titles_only.result', 'w') as f:
        for item in doc:
            f.write("%s\n" % item)    

            
def extract_fr(doc_with_list):
    lines = []
    doc = []

    f = open(doc_with_list, "r")
    lines = lines + f.read().splitlines()

    f_text = []
    f_id = []

    for i in os.listdir('fr94'):
        for j in os.listdir('fr94'+'/'+i):
            f = open('fr94/'+i+'/'+j)
            f_text.append(f.read().splitlines())
            f_id.append(j[:5])


    for i in lines:
        for j in range(len(f_text[f_id.index(i[:5].lower())])):
            if i in f_text[f_id.index(i[:5].lower())][j]:
                p=j-1
                while '</DOC>' not in f_text[f_id.index(i[:5].lower())][p]:
                    doc.append(f_text[f_id.index(i[:5].lower())][p])
                    p+=1
                doc.append('</DOC>')
                break


    with open('fr_titles_only.result', 'w') as f:
        for item in doc:
            f.write("%s\n" % item)

def extract_fb(doc_with_list):
    lines = []
    doc = []

    f = open(doc_with_list, "r")
    lines = lines + f.read().splitlines()

    text_3 = []
    text_4 = []
    fid_3 = []
    fid_4 = []

    for i in os.listdir('fbis'):
        f = open('fbis/'+i)
        if i[2]=='3':
            temp = f.read().splitlines()
            text_3.append(temp)
            p=0
            while '<DOCNO>' not in temp[p]:
                p+=1
            fid_3.append(int(temp[p].split(' ')[1].split('-')[1]))
        else:
            temp = f.read().splitlines()
            text_4.append(temp)
            p=0
            while '<DOCNO>' not in temp[p]:
                p+=1
            fid_4.append(int(temp[p].split(' ')[1].split('-')[1]))

    for i in lines:
        if i[4]=='3':
            pid=0
            while True:
                if int(i.split('-')[1])<fid_3[pid]:
                    pid = pid-1
                    break
                pid+=1
            for j in range(len(text_3[pid])):
                if i in text_3[pid][j]:
                    p=j-1
                    while '</DOC>' not in text_3[pid][p]:
                        doc.append(text_3[pid][p])
                        p+=1
                    doc.append('</DOC>')
                    break
        else:
            pid=0
            while True:
                if int(i.split('-')[1])<fid_4[pid]:
                    pid = pid-1
                    break
                pid+=1
            for j in range(len(text_4[pid])):
                if i in text_4[pid][j]:
                    p=j-1
                    while '</DOC>' not in text_4[pid][p]:
                        doc.append(text_4[pid][p])
                        p+=1
                    doc.append('</DOC>')
                    break

    with open('fb_titles_only.result', 'w') as f:
        for item in doc:
            f.write("%s\n" % item)
```
