```python
import os

lines = []
for i in os.listdir('topics'):
    f = open('topics/' + i, "r")
    lines = lines + f.read().splitlines()
```

```python
lines
```

```python
'<title>' in lines[3]

titles = []
desc = []
nar = []

i=0
while i < len(lines):
    if '<title>' in lines[i]:
        titles.append(lines[i][7:])
    if '<desc>' in lines[i]:
        i=i+1
        temp = ''
        while '<narr>' not in lines[i]:
            temp = temp + ' ' +lines[i]
            i=i+1
        desc.append(temp)
    if '<narr>' in lines[i]:
        i=i+1
        temp = ''
        while '</top>' not in lines[i]:
            temp = temp + lines[i]
            i=i+1
        nar.append(temp)
    i+=1
    
```

```python
for i in range(0,len(titles)):
    titles[i]=titles[i][1:-1]
    desc[i]=desc[i][1:-1]
for i in range(0,len(titles)):
    for j in ['.'  ,',' , '!', '?', '-', '(', ')', "'", '"', '/', '[' , ']', '{', '}' , '+', '%', '#', '@','$','^','&','*','_','|','  ']:
        titles[i]=titles[i].replace(j,' ')
        desc[i]=desc[i].replace(j,' ')
        nar[i]=nar[i].replace(j,' ')
```

```python
final = ['<parameters>','<index></index>','<rule>method:dirichlet,mu:1000</rule>','<count>1000</count>','<trecFormat>true</trecFormat>']
num = 301
for i in range(0,len(titles)):
    final = final + ['<query> <type>indri</type> <number>'+str(num)+'</number> <text>'+titles[i]+' '+desc[i]+'</text> </query>']
    num+=1
final = final + ['</parameters>']

with open('IndriRunQuerry.queries.file.301-450-titles-descs.ERGASIA', 'w') as f:
    for item in final:
        f.write("%s\n" % item)
```
