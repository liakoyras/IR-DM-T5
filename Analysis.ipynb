```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sw = pd.read_csv('Results/stopwords.csv')
n_sw = pd.read_csv('Results/no_stopwords.csv')
ini_titles=[i * 100 for i in [0.1186,0.1608,0.326]]
ini_titles_descs=[i * 100 for i in [0.2239,0.2419,0.3918]]
ini_titles_descs_nar=[i * 100 for i in [0.2162,0.2483,0.4130]]
names = ['titles','titles_descs','titles_descs_nar']
```

```python
fig, ax = plt.subplots(figsize=(15,8))

stopwords = ax.bar(np.arange(4),sw['map'].apply(lambda x: x*100).to_list(),0.35,label='With Stopwords',color='steelblue')
n_stopwords = ax.bar(np.arange(4)+0.35,n_sw['map'].apply(lambda x: x*100).to_list(),0.35,label='Without Stopwords',color='khaki')
ini = ax.bar(np.arange(4)+0.35/2,[ini_titles[0],ini_titles[0],ini_titles_descs[0],ini_titles_descs[0]],0.7,label='Initial Queries',alpha = 0.5,color='maroon')

ax.set_title('MAP score on query files run w/ and w/o stopwords')
ax.set_xlabel('Query File')
ax.set_ylabel('MAP score (%)')

ax.legend()
ax.set_xticks(np.arange(4) + 0.35 / 2)
ax.set_xticklabels(sw['ID'].to_list())
plt.setp(ax.xaxis.get_majorticklabels(), rotation=45)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,8))

stopwords = ax.bar(np.arange(4),sw['r-prec'].apply(lambda x: x*100).to_list(),0.35,label='With Stopwords',color='steelblue')
n_stopwords = ax.bar(np.arange(4)+0.35,n_sw['r-prec'].apply(lambda x: x*100).to_list(),0.35,label='Without Stopwords',color='khaki')
ini = ax.bar(np.arange(4)+0.35/2,[ini_titles[1],ini_titles[1],ini_titles_descs[1],ini_titles_descs[1]],0.7,label='Initial Queries',alpha = 0.5,color='maroon')

ax.set_title('R-Precision score on query files run w/ and w/o stopwords')
ax.set_xlabel('Query File')
ax.set_ylabel('R-Precision score (%)')

ax.legend()
ax.set_xticks(np.arange(4) + 0.35 / 2)
ax.set_xticklabels(sw['ID'].to_list())
plt.setp(ax.xaxis.get_majorticklabels(), rotation=45)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,8))

stopwords = ax.bar(np.arange(4),sw['p10'].apply(lambda x: x*100).to_list(),0.35,label='With Stopwords',color='steelblue')
n_stopwords = ax.bar(np.arange(4)+0.35,n_sw['p10'].apply(lambda x: x*100).to_list(),0.35,label='Without Stopwords',color='khaki')
ini = ax.bar(np.arange(4)+0.35/2,[ini_titles[2],ini_titles[2],ini_titles_descs[2],ini_titles_descs[2]],0.7,label='Initial Queries',alpha = 0.5,color='maroon')

ax.set_title('P@10 score on query files run w/ and w/o stopwords')
ax.set_xlabel('Query File')
ax.set_ylabel('P@10 score (%)')

ax.legend()
ax.set_xticks(np.arange(4) + 0.35 / 2)
ax.set_xticklabels(sw['ID'].to_list())
plt.setp(ax.xaxis.get_majorticklabels(), rotation=45)

plt.show()
```
